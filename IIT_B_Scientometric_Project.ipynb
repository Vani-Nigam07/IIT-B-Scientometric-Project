```python
import pandas as pd
from tabulate import tabulate
# Load the dataset


#df= pd.read_csv('/content/csv10.csv', encoding='unicode_escape')
df=pd.read_csv('/content/csv2.csv', encoding='unicode_escape')

# Print the DataFrame using tabulate
#print(tabulate(df.head(), headers='keys', tablefmt='psql'))

```

```python
df['text'] = df['TI'] + ' ' + df['AB']


#print(tabulate(df.head(), headers='keys', tablefmt='psql'))


```

```python
import re

# Combine title and abstract into a single text column
df['text'] = df['TI'] + ' ' + df['AB']


# Function to preprocess text
def preprocess_text(text):
    if not isinstance(text, str):
        return ''
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text
# Apply the function to the text column
df['text'] = df['text'].apply(preprocess_text)

# Display the first few rows to verify preprocessing
#print(tabulate(df.head(), headers='keys', tablefmt='psql'))

```

# to convert into string

```python
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords if not already done
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
english_stopwords = stopwords.words('english')
custom= ['could', 'also','conditions','time']
# For example, remove 'not' from stopwords
english_stopwords.extend(custom)


# Function to remove stopwords
def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in english_stopwords])

# Apply the function to the text column
df['text'] = df['text'].apply(remove_stopwords)

# Display the first few rows to verify stopword removal
print(tabulate(df[['text']].head()))
```

```python
import re

# Refined list of common catalyst names/patterns
catalyst_patterns = [
    #r'\bPt',    # Matches Pt
    r'Pt[A-Z][a-z]*\b',    # Matches Pt-based compounds (e.g., PtCl4)
    r'\bNi\b',    # Matches Ni
    r'\bNi[A-Z][a-z]*\b',    # Matches Ni-based compounds (e.g., NiCl2)
    r'\bPalladium\b',  # Matches Palladium
    r'\bIron\b',     # Matches Iron
    r'Fe\b',    # Matches Fe
    r'\bFe[A-Z][a-z]*\b',    # Matches Fe-based compounds (e.g., FeCl3)
                            # Matches Co
    r'\bCo[A-Z][a-z]*\b',    # Matches Co-based compounds (e.g., CoCl2)
    r'\bCopper\b',   # Matches Copper
    r'\bCu\b',    # Matches Cu
    r'\bCu[A-Z][a-z]*\b',  # Matches Cu-based compounds (e.g., CuO)
    r'\bZr[A-Z][a-z]*\b',
    r'\bZeolite\b',  # Matches Zeolite
    r'\bTitania\b',  # Matches Titania (common name for TiO2)
    r'\bTi\b',    # Matches Ti
    r'\bTi[A-Z][a-z]*\b',    # Matches Ti-based compounds (e.g., TiO2)
    r'\bBoro[a-z]*\b',  # Matches Boron-based compounds (e.g., Boron, Borosilicate)
    r'\bCarbon\b',   # Matches Carbon
    r'\bCarbon[A-Z][a-z]*\b',  # Matches Carbon-based compounds (e.g., Carbon nanotube)
    r'\bRutile\b',  # Matches Rutile (common name for a form of TiO2)
    r'doped\b',  # Matches doped (used in context of doping with another element)
    r'alloy\b',   # Matches alloy (used in context of metal alloys)
    r'\brare',
    r'\bmetal\b'

]

def extract_catalysts(text, patterns):
    catalysts = []
    for pattern in patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        catalysts.extend(matches)
    return list(set(catalysts))


df['catalysts'] = df['text'].apply(lambda x: extract_catalysts(x, catalyst_patterns))

# Display the DataFrame with extracted catalysts
print(tabulate(df[['text', 'catalysts']][:50]))
```

# spaCy Model

```python
# Load the spaCy model
import spacy
nlp = spacy.load('en_core_web_sm')

# Function to extract entities
def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text for ent in doc.ents if ent.label_ in ['ORG', 'PRODUCT']]  # Customize label conditions if necessary
    return entities

# Apply the function to the text column
df['entities'] = df['catalysts'].apply(extract_entities)

# Display the first few rows with extracted entities
if 'entities' in df.columns:
  print(tabulate(df[['catalysts', 'entities']].head()))

```

```python
from nltk import word_tokenize
import nltk
nltk.download('punkt')  # Ensure that you have the tokenizer models

# Sample text
for i in range(len(df)):
  text = df['text'][i]
  tokens = word_tokenize(text)
  filtered_tokens = []
  for i in range(len(tokens)):
      current_token = tokens[i].lower()
      preceding_word = tokens[i-1].lower() if i > 0 else ''

    # Check if the current token is a custom stopword with the preceding word condition
      if preceding_word in custom_stopwords and current_token in custom_stopwords[preceding_word]:
          continue
      elif '' in custom_stopwords and current_token in custom_stopwords['']:
          continue
      else:
          filtered_tokens.append(tokens[i])

# Print the filtered tokens
print(filtered_tokens)
```
